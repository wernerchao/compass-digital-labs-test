```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import pylab
```

```python
data = pd.read_csv('../data/siteB.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
import datetime
```

```python
print data['Date_Name'].head()
data['Date_Name'] = data['Date_Name'].apply(lambda d: datetime.datetime.strptime(d, '%Y-%m-%d'))
print data.info()
```

```python
temp = data.groupby(by=['Date_Name','subCategory'])['sales'].sum()
print temp.head()
```

```python
temp = pd.DataFrame(data={'sales': temp})
print temp.head()
temp.reset_index(inplace=True)
print temp.head()
```

```python
plt.rcParams['figure.figsize'] = 16,8
cat_A = temp[temp['subCategory'] == 'A']
plt.plot(cat_A['Date_Name'], cat_A['sales'], '-o')
plt.ylabel('$ Sales')
plt.title('Sales vs Time')
```

```python
plt.rcParams['figure.figsize'] = 8, 4
_ = plt.boxplot(cat_A['sales'])
_ = plt.xticks([1], ['A'])
_ = plt.ylabel('$ Sales')
```

```python
_ = plt.hist(cat_A['sales'], bins=50)
_ = plt.xlabel('$ Sales')
_ = plt.ylabel('Count')
_ = plt.title('Histogram of $ Sales')
```
